```python
#just to know everything well
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target
feature_names = iris.feature_names
class_names = iris.target_names

# Function to train and plot a tree
def show_tree(criterion, max_depth, min_samples_split, min_samples_leaf, max_features, title):
    clf = DecisionTreeClassifier(
        criterion=criterion,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        max_features=max_features,
        random_state=42
    )
    clf.fit(X, y)

    plt.figure(figsize=(12,6))
    plot_tree(clf, feature_names=feature_names, class_names=class_names, filled=True, fontsize=8)
    plt.title(title, fontsize=14)
    plt.show()

# 1. criterion
show_tree("gini", max_depth=3, min_samples_split=2, min_samples_leaf=1, max_features=None,
          title="Criterion = Gini (default)")

show_tree("entropy", max_depth=3, min_samples_split=2, min_samples_leaf=1, max_features=None,
          title="Criterion = Entropy (information gain)")

# 2. max_depth
show_tree("gini", max_depth=2, min_samples_split=2, min_samples_leaf=1, max_features=None,
          title="Max Depth = 2 (shallow tree)")

show_tree("gini", max_depth=5, min_samples_split=2, min_samples_leaf=1, max_features=None,
          title="Max Depth = 5 (deeper tree)")

# 3. min_samples_split
show_tree("gini", max_depth=4, min_samples_split=10, min_samples_leaf=1, max_features=None,
          title="Min Samples Split = 10 (less branching)")

# 4. min_samples_leaf
show_tree("gini", max_depth=4, min_samples_split=2, min_samples_leaf=5, max_features=None,
          title="Min Samples Leaf = 5 (each leaf has â‰¥5 samples)")

# 5. max_features
show_tree("gini", max_depth=4, min_samples_split=2, min_samples_leaf=1, max_features=2,
          title="Max Features = 2 (randomly selects 2 features at each split)")
```

```python
#45 Train a Bagging Regressor with different levels of bootstrap samples and compare performance.
x,y=make_regression(n_samples=500,n_features=20)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
bootstrap_samples = [0.5, 0.7, 1.0]
bootstrap_options = [True, False]     # with and without replacement

for bootstrap in bootstrap_options:
  for frac in bootstrap_samples:
        model = BaggingRegressor(
            estimator=DecisionTreeRegressor(),
            n_estimators=100,
            max_samples=frac,
            bootstrap=bootstrap,# controls bootstrap sample size
            random_state=42
        )
        model.fit(x_train,y_train)
        y_pred = model.predict(x_test)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        print(f"  max_samples={frac} | RMSE={rmse:.3f}")

```
